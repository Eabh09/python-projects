```python
stock=yf.Ticker("AAPL").history(period='1d',interval='1m')
x= list()
for i in range(stock.shape[0]):
    
clf=MinMaxScaler()
x=clf.fit_transform(x.reshape(-1,1))
y=clf.fit_transform(y.reshape(-1,1
))

train_labels=int(stock.shape[0]*0.7)
x_train=x[:train_labels]
y_train=y[:train_labels]
x_train
```

```python
from tensorflow.keras.layers import LSTM,Dense
from tensorflow.keras import Sequential
from tensorflow.keras.metrics import BinaryAccuracy
from keras.layers.core import Dense, Activation, Dropout
import tensorflow as tf
model=Sequential()
model.add(tf.keras.layers.Input(shape=(1,1)))
model.add(Dense(1))
model.compile(optimizer="Adam", loss="mse",metrics=[tf.keras.metrics.BinaryAccuracy()])
model.fit(x_train.reshape(-1,1,1),y_train.reshape(-1,1,1),batch_size=25,epochs=1000,use_multiprocessing=True)
```

```python
predict=model.predict(x[train_labels:])

pr=clf.inverse_transform(predict.reshape(-1,1))
pr
```
